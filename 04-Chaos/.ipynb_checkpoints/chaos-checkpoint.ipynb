```python
#Imports
import numpy as np                    #numpy for math
import matplotlib.pyplot as plt       #matplotlib for plotting
import cartopy.crs as ccrs            #cartopy for plotting on map
import cartopy.feature as cfeature    #cartopy basic shapefiles
```

```python
#Define the constants
a = 0.25
b = 4.0
F = 8.0
G = 1.0
time = (4/(24*5))
```

```python
#Define the equations (put into a function)

def time_eqn(x,y,z):
    
    """
    Created by Lorenz(1984, 1990, 2006)
    "x" represents the strength of a circumpolar westerly current
    "y" and "z" denote the cosine and sine phases of a superposed chain of large-scale waves
    "t" is a normalized time such that 1 normalized time unit equals 5 days
    "a", "b", "F", and "G" are constants
    """
    dx = (-(y**2) - (z**2) - (a*x) + (a*F))
    time_step_x = x + (time*(dx))
    
    dy = ((x*y) - (b*x*z) - y + G)
    time_step_y = y + (time*(dy))
    
    dz = ((b*x*y) + (x*z) - z)
    time_step_z = z + (time*(dz))
    
    return ([time_step_x, time_step_y, time_step_z]) #making it into a list
```

```python
#This is for Simulation #1
first_conditions_x = np.empty([720])
first_conditions_y = np.empty([720])
first_conditions_z = np.empty([720])

#Set the first values in each list to the first set of initial conditions
first_conditions_x[0] = 2.5
first_conditions_y[0] = 1.0
first_conditions_z[0] = 0.0

#Create a loop that will start from 1 and go until 720 (since our "0" is the initial conditions)

for i in range(1, 720):
    run = time_eqn(first_conditions_x[i-1], first_conditions_y[i-1], first_conditions_z[i-1]) #plug into the function
    first_conditions_x[i] = run[0] #grab just the x
    first_conditions_y[i] = run[1] #grab just the y
    first_conditions_z[i] = run[2] #grab just the z
```

```python
#Plot the first set of initial conditions (X)
plt.plot(first_conditions_x[:], color = 'indigo')
plt.ylabel("Strength of Circumpolar Westerly Current")
plt.xlabel("Time in Days")
plt.title("Simulation #1 (X)")
```

```python
#Plot the first set of initial conditions (Y)
plt.plot(first_conditions_y[:], color = 'lime')
plt.ylabel("Phases of Superposed Chain of Large Scale Waves")
plt.xlabel("Time in Days")
plt.title("Simulation #1 (Y)")
```

```python
#Plot the first set of initial conditions (Z)
plt.plot(first_conditions_z[:], color = 'mediumvioletred')
plt.ylabel("Strength of Circumpolar Westerly Current")
plt.xlabel("Time in Days")
plt.title("Simulation #1 (Z)")
```

```python
#Doing the same process as above, just for the second simulation

#This is for Simulation #2
second_conditions_x = np.empty([720])
second_conditions_y = np.empty([720])
second_conditions_z = np.empty([720])

#Set the first values in each list to the 2nd set of initial conditions
second_conditions_x[0] = 2.49
second_conditions_y[0] = 1.0
second_conditions_z[0] = 0.0

#Create a loop that will start from 1 and go until 720 (since our "0" is the initial conditions)

for i in range(1, 720):
    run_2 = time_eqn(second_conditions_x[i-1], second_conditions_y[i-1], second_conditions_z[i-1])
    second_conditions_x[i] = run_2[0]
    second_conditions_y[i] = run_2[1]
    second_conditions_z[i] = run_2[2]
```

```python
#Plot the second set of initial conditions (X)
plt.plot(second_conditions_x[:], color = 'mediumorchid')
plt.ylabel("Strength of Circumpolar Westerly Current")
plt.xlabel("Time in Days")
plt.title("Simulation #2 (X)")
```

```python
#Plot the second set of initial conditions (Y)
plt.plot(second_conditions_y[:], color = 'forestgreen')
plt.ylabel("Phases of Superposed Chain of Large Scale Waves")
plt.xlabel("Time in Days")
plt.title("Simulation #2 (Y)")
```

```python
#Plot the second set of initial conditions (Z)
plt.plot(second_conditions_z[:], color = 'hotpink')
plt.ylabel("Phases of Superposed Chain of Large Scale Waves")
plt.xlabel("Time in Days")
plt.title("Simulation #2 (Z)")
```

```python
#Create subplots for X first and second simulations
fig, axs = plt.subplots(2)
plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.85,wspace=0.9,hspace=0.7)
fig.suptitle('First & Second Simulations for X')
axs[0].plot(first_conditions_x[:], color = 'indigo')
axs[0].set_title('Simulation #1 (X)')
axs[0].set_xlabel('Time in Days')
axs[1].plot(second_conditions_x[:], color = 'mediumorchid')
axs[1].set_title('Simulation #2 (X)')
axs[1].set_xlabel('Time in Days')
```

```python
#Plot both first and second simulation (X) together on one plot
plt.plot(first_conditions_x[:], color = 'indigo', label = 'First Simulation')
plt.plot(second_conditions_x[:], color = 'mediumorchid', label = 'Second Simulation')
plt.xlabel("Time in Days")
plt.title("First and Second Simulation (X)")
plt.legend(loc = 'lower right')
```

```python
#Create subplots for y first and second conditions
fig, axs = plt.subplots(2)
plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.85,wspace=0.9,hspace=0.7)
fig.suptitle('First & Second Simulation for Y')
axs[0].plot(first_conditions_y[:], color = 'lime')
axs[0].set_title('First Simulation (Y)')
axs[0].set_xlabel('Time in Days')
axs[1].plot(second_conditions_y[:], color = 'forestgreen')
axs[1].set_title('Second Simulation (Y)')
axs[1].set_xlabel('Time in Days')
```

```python
#Plot both first and second simulation (Y) together on one plot
plt.plot(first_conditions_y[:], color = 'lime', label = 'First Simulation')
plt.plot(second_conditions_y[:], color = 'forestgreen', label = 'Second Simulation')
plt.xlabel("Time in Days")
plt.title("First and Second Simulation (Y)")
plt.legend(loc = 'lower right')
```

```python
#Create subplots for Z first and second conditions
fig, axs = plt.subplots(2)
plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.85,wspace=0.9,hspace=0.7)
fig.suptitle('First & Second Simulation for Z')
axs[0].plot(first_conditions_z[:], color = 'mediumvioletred')
axs[0].set_title('First Simulation (Z)')
axs[0].set_xlabel('Time in Days')
axs[1].plot(second_conditions_z[:], color = 'hotpink')
axs[1].set_title('Second Simulation (Z)')
axs[1].set_xlabel('Time in Days')
```

```python
#Plot both first and second simulation (Z) together on one plot
plt.plot(first_conditions_z[:], color = 'mediumvioletred', label = 'First Simulation')
plt.plot(second_conditions_z[:], color = 'hotpink', label = 'Second Simulation')
plt.xlabel("Time in Days")
plt.title("First and Second Simulation (Z)")
plt.legend(loc = 'lower right')
```
